# Normalize Stock Prices as an input to a statistic analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import preprocessing
#%matplotlib notebook

dfg = pd.read_excel(r'F:\commodity data last 2 yr\Gold Futures Historical Data.xlsx',parse_dates = True,index_col = 0)
dfs = pd.read_excel(r'F:\commodity data last 2 yr\Silver Futures Historical Data.xlsx',parse_dates = True,index_col = 0)
dfga = pd.read_excel(r'F:\commodity data last 2 yr\Natural Gas Futures Historical Data.xlsx',parse_dates = True,index_col = 0)
dfc = pd.read_excel(r'F:\commodity data last 2 yr\Crude Oil WTI Futures Historical Data.xlsx',parse_dates = True,index_col = 0)
```

CONCATENATE  DATA

```python

tickers = ['dfg','dfs','dfga','dfc']

stocks = pd.concat([dfg,dfs,dfga,dfc],axis=1,keys=tickers)
```

```python
stocks.columns.names = ['comodity', 'stocks info']
```

```python
final = stocks.xs(key= 'Price', axis=1,level = 'stocks info')
```

```python
dividorG = final.iloc[0,0]
dividorG
final=final.copy()
final['dfg_norm'] = (100*final['dfg'])/dividorG
```

```python
dividorS = final.iloc[0,1]
dividorS
final=final.copy()
final['dfs_norm'] = (100*final['dfs'])/dividorS
```

```python
dividorGA = final.iloc[0,2]
dividorGA
final=final.copy()
final['dfga_norm'] = (100*final['dfga'])/dividorGA
```

```python
dividorC = final.iloc[0,3]
dividorC
final=final.copy()
final['dfc_norm'] = (100*final['dfc'])/dividorC
```

```python
plt.figure(figsize=(15,5))
plt.grid(True)
plt.plot(final['dfg_norm'],label="GLD NORMALIGE")
plt.plot(final['dfs_norm'],label="SILVER NORMALIGE")
plt.plot(final['dfga_norm'],label="GAS NORMALIGE")
plt.plot(final['dfc_norm'],label="CRUDE NORMALIGE")


plt.legend(loc = 2)
```
